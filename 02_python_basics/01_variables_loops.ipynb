```python
# 02_variables_loops.ipynb

# Variables & Data Types
name = "Rahul"
age = 28
is_student = True
scores = [85, 90, 78, 92]

print("Name:", name)
print("Age:", age)
print("Is Student?", is_student)
print("Scores:", scores)

# Loops
print("\nðŸ“š Scores with loop:")
for i, score in enumerate(scores, 1):
    print(f"Test {i}: {score}")

# Conditionals
print("\nðŸŽ¯ Performance:")
avg = sum(scores) / len(scores)
if avg >= 90:
    print("Grade: A")
elif avg >= 80:
    print("Grade: B")
else:
    print("Grade: C")

# Function
def greet(name):
    return f"Hello, {name}! Welcome to Data Science."

print("\nðŸ‘‹", greet(name))
```
