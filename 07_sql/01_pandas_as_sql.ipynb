```python
# 07_sql/01_pandas_as_sql.ipynb
import pandas as pd

# Load customer data
df = pd.read_csv("https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv")

print("üîç SQL Operations using Pandas\n")

# 1. SELECT
print("1Ô∏è‚É£ SELECT customerID, gender, Churn")
selected = df[['customerID', 'gender', 'Churn']]
print(selected.head(3))

# 2. WHERE
print("\n2Ô∏è‚É£ WHERE Churn = 'Yes'")
churned = df[df['Churn'] == 'Yes']
print(f"Total churned customers: {len(churned)}")

# 3. GROUP BY + COUNT
print("\n3Ô∏è‚É£ GROUP BY gender, COUNT(Churn)")
churn_by_gender = df[df['Churn'] == 'Yes'].groupby('gender').size()
print(churn_by_gender)

# 4. ORDER BY
print("\n4Ô∏è‚É£ ORDER BY MonthlyCharges DESC (Top 5)")
top_customers = df.nlargest(5, 'MonthlyCharges')[['customerID', 'MonthlyCharges']]
print(top_customers)

# 5. JOIN (Merge)
print("\n5Ô∏è‚É£ JOIN: Add region info")
regions = pd.DataFrame({
    'customerID': df['customerID'].head(10),
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South']
})
joined = pd.merge(df.head(10), regions, on='customerID', how='left')
print(joined[['customerID', 'Churn', 'region']].head())
```
