```python

# 12_feature_engineering/01_encoding_scaling.ipynb
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

# Load data
df = pd.read_csv("https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv")

# ✅ FIX: Handle TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Create new feature
df['AvgMonthlySpend'] = df['TotalCharges'] / (df['tenure'] + 1)  # +1 to avoid division by zero
df['AvgMonthlySpend'] = df['AvgMonthlySpend'].fillna(df['MonthlyCharges'])

# Separate features
numerical_features = ['tenure', 'MonthlyCharges', 'AvgMonthlySpend']
categorical_features = ['gender', 'Partner', 'Dependents']

# One-Hot Encoding for categorical
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ]
)

X_processed = preprocessor.fit_transform(df)

print("✅ Feature Engineering Complete!")
print(f"Original features: {len(numerical_features) + len(categorical_features)}")
print(f"Processed features: {X_processed.shape[1]} (after One-Hot Encoding)")
print(f"Sample of first row:\n{X_processed[0]}")
```
