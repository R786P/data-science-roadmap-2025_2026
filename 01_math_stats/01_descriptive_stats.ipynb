```python
# 01_descriptive_stats.ipynb
import pandas as pd
from scipy import stats

# Load telecom customer data
df = pd.read_csv("https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv")

# Descriptive Statistics
print("ðŸ“Š Descriptive Statistics")
print("Mean Monthly Charges:", round(df['MonthlyCharges'].mean(), 2))
print("Median Monthly Charges:", round(df['MonthlyCharges'].median(), 2))
print("Standard Deviation:", round(df['MonthlyCharges'].std(), 2))

# Correlation
print("\nðŸ”— Correlation (tenure vs MonthlyCharges):")
corr = df[['tenure', 'MonthlyCharges']].corr().iloc[0,1]
print(f"{corr:.3f}")

# Hypothesis Testing
print("\nðŸ§ª Hypothesis Test: Do churned customers pay more?")
churn_yes = df[df['Churn'] == 'Yes']['MonthlyCharges']
churn_no = df[df['Churn'] == 'No']['MonthlyCharges']
t_stat, p_val = stats.ttest_ind(churn_yes, churn_no, equal_var=False)
print(f"p-value: {p_val:.4f}")
if p_val < 0.05:
    print("âœ… Yes! Churned customers pay significantly more.")
else:
    print("âŒ No significant difference.")
```
