```python
# 06_eda/01_univariate_bivariate.ipynb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and prep data
df = pd.read_csv("https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv")

# Univariate Analysis
print("ðŸ“ˆ Univariate Analysis")

# Churn distribution
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Churn', palette='Set1')
plt.title('Churn Distribution')

# Tenure histogram
plt.subplot(1, 2, 2)
sns.histplot(df['tenure'], kde=True, color='green')
plt.title('Tenure Distribution')
plt.tight_layout()
plt.show()

# Bivariate Analysis
print("\nðŸ”— Bivariate Analysis")

# Churn vs Gender
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='gender', hue='Churn', palette='Set2')
plt.title('Churn by Gender')

# Churn vs Tenure
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Churn', y='tenure', palette='Set3')
plt.title('Tenure vs Churn')
plt.tight_layout()
plt.show()

# Summary stats by Churn
print("\nðŸ“Š Summary Stats by Churn:")
print(df.groupby('Churn')['MonthlyCharges'].describe())
```
